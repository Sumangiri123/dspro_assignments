1. Define a doubly linked list

```python
# Defining class node

class Node:
    def __init__(self,Data,Prev=None,Next=None):
        self.__Data = Data
        self.__Prev = Prev
        self.__Next = Next
    
    def setData(self,Data):
        self.__Data = Data
        
    def getData(self):
        return self.__Data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next
    
    def setPrev(self,Prev):
        self.__Prev = Prev
        
    def getPrev(self):
        return self.__Prev
        
# Creating list nodes

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)

#Creating Doubly linked list

head.setNext(node1)
node1.setNext(node2)
node1.setPrev(head)
node2.setNext(node3)
node2.setPrev(node1)
node3.setNext(node4)
node3.setPrev(node2)
node4.setPrev(node3)

# Traversing the doubly linked list

def traverse(head):
    temp = head
    while (temp):
        print(temp.getData(),end="<->")
        temp = temp.getNext()

# Calling the function for traversing the list  
print("Doubly Linked List : ",end = "")
traverse(head)
```

2. Write a function to reverse a linked list in_place

```python
# Creating class Node
class Node:
    def __init__(self,Data,Next=None):
        self.__Data = Data
        self.__Next = Next
    
    def setData(self,Data):
        self.__Data = Data
        
    def getData(self):
        return self.__Data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next

# Creating list nodes

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)    
        
    
# Creating a linked list

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)

# Traversing the linked list

def traverse(head):
    temp = head
    while (temp):
        print(temp.getData(),end="->")
        temp = temp.getNext()
    

# Calling the function for traversing the list        
print("Original Linked list : ")
traverse(head)

# Reversing the linked list

def reverse_LL(head):
    
    prev_node = None
    current_node = head
    
    while current_node is not None:
        next_node = current_node.getNext()  # Store the next node
        current_node.setNext(prev_node)     # Reverse the pointer
        
        # Move to the next pair of nodes
        prev_node = current_node
        current_node = next_node
    
    # At the end, the prev_node will be the new head
    head = prev_node
    
    return head

print()
new_head = reverse_LL(head)
print("Reversed Linked List :")
traverse(new_head)
```

3. Detect cycle in a linked list.

```python
class Node:
    
    def __init__(self,data,Next=None):
        self.__data = data
        self.__Next = Next
        
    def setData(self,data):
        self.__data = data
        
    def getData(self):
        return self.__data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next
    
# Creating list Nodes
head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)

# Creating linked list with cycle 
head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node2)

def check_cycle(head): # Checks whether linked list has a cycle or not
    slow = head
    fast = head
    while (fast and fast.getNext()):
        
        slow = slow.getNext()
        fast = fast.getNext().getNext()
            
        if slow.getData() == fast.getData():
            return True
        
    return False         

check_cycle(head)
         
```

4. Merge Two sorted Linked List .
1->3->5->7->null and 2->4->6->8->null should be merged to make 1->2->3->4->5->6->7->8

```python
class Node:
    
    def __init__(self,data,Next=None):
        self.__data = data
        self.__Next = Next
        
    def setData(self,data):
        self.__data = data
        
    def getData(self):
        return self.__data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next
    
# Creating list Nodes of List 1
head_1 = Node(1)
node1 = Node(3)
node2 = Node(5)
node3 = Node(7)

# Creating linked list 1
head_1.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)


# Creating list Nodes of List 2
head_2 = Node(2)
node_1 = Node(4)
node_2 = Node(6)
node_3 = Node(8)

# Creating linked list 2
head_2.setNext(node_1)
node_1.setNext(node_2)
node_2.setNext(node_3)


def traverse(head):
    temp = head
    while (temp):
        print(temp.getData(),end="->")
        temp = temp.getNext()
    
print("Linked List 1 : ",end = "")
traverse(head_1)

print()

print("Linked List 2 : ",end = "")
traverse(head_2)

print()

print("Merged sorted List : ",end = "")

def merge_sorted_lists(head1, head2):
    # Create a dummy node to serve as the head of the merged list
    dummy = Node(0)
    current = dummy
    # Traverse both lists
    while head1 and head2:
        if head1.getData() < head2.getData():
            current.setNext(head1)
            head1 = head1.getNext()
        else:
            current.setNext(head2)
            head2 = head2.getNext()
            
        current = current.getNext()

    # Append remaining nodes from either list
    if head1:
        current.setNext(head1)
    elif head2:
        current.setNext(head2)

    # Return the head of the merged list (skipping the dummy node)
    return dummy.getNext()

merge = merge_sorted_lists(head_1,head_2)
traverse(merge)
```

5. Write a function to remove nth node from the end in a linked list .
1->2->3->4->5->6, removing 2nd node from end will return 1->2->3->4->6

```python
class Node:
    
    def __init__(self,data,Next=None):
        self.__data = data
        self.__Next = Next
        
    def setData(self,data):
        self.__data = data
        
    def getData(self):
        return self.__data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next
    
# Creating list Nodes
head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(5)
node5 = Node(6)

# Creating linked list with cycle 
head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)

def traverse(head):
    temp = head 
    while temp:
        print(temp.getData(),end= "->")
        temp = temp.getNext()

print("Given Linked List : ",end = "")
traverse(head)

print()

def length(head):
    l = 0
    temp = head
    while temp:
        l += 1
        temp = temp.getNext()

    return l 

def remove_node(head,n): # deleting nth node from end 
    k = length(head) - n + 1 # kth node from starting
    
    # checking if k is valid or not
    if k < 0 or k > length(head):
        print("Invalid Input")
        return head
    
    if k == 0: # remove the begining node
        head = head.getNext()
    else:
        # traversing to the previous node
        prev = head
        i = 1
        while i < k-1:
            prev = prev.getNext()
            i += 1

    prev.setNext(prev.getNext().getNext())
    
    return head

# Deleting 2nd node from end
newhead = remove_node(head,2)
print("After deleting desired Node : ",end = "")
traverse(newhead)
```

6. Remove duplicates from a sorted linked list
    1->2->3->3->4->4->4->5 should be changed to 1->2->3->4->5

```python
# Creating class Node
class Node:
    def __init__(self,Data,Next=None):
        self.__Data = Data
        self.__Next = Next
    
    def setData(self,Data):
        self.__Data = Data
        
    def getData(self):
        return self.__Data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next

# Creating list nodes

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(3)
node4 = Node(4)
node5 = Node(4)
node6 = Node(4)
node7 = Node(5)
        
    
# Creating a linked list

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)
node5.setNext(node6)
node6.setNext(node7)

# Traversing the linked list

def traverse(head):
    temp = head
    while (temp):
        print(temp.getData(),end="->")
        temp = temp.getNext()
    

# Calling the function for traversing the list

print("Original Linked list : ",end = "")
traverse(head)

print()
print()

# Removing Duplicate elements from linked list

def removeDuplicates(head):
    current = head
    while current and current.getNext():
        if current.getData() == current.getNext().getData():
            # Skip the duplicate node by pointing the current node's next to the next of the next node
            current.setNext(current.getNext().getNext())
        else:
            current = current.getNext()
    return head


# Calling the function for traversing the roated linked list        
print("Modified Linked list : ",end = "")
newhead = removeDuplicates(head)
traverse(newhead)
```

7. Find the intersection of the two linked lists
    1->2->3->4->8->6->9 ; 5->1->6->7 ; intersection 1->6

```python
class Node:
    
    def __init__(self,data,Next=None):
        self.__data = data
        self.__Next = Next
        
    def setData(self,data):
        self.__data = data
        
    def getData(self):
        return self.__data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next
    
# Creating list Nodes of List 1
head_1 = Node(1)
node1 = Node(3)
node2 = Node(4)
node3 = Node(8)
node4 = Node(6)
node5 = Node(9)

# Creating linked list 1
head_1.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)

# Creating list Nodes of List 2
head_2 = Node(5)
node_1 = Node(1)
node_2 = Node(6)
node_3 = Node(7)

# Creating linked list 2
head_2.setNext(node_1)
node_1.setNext(node_2)
node_2.setNext(node_3)

def traverse(head):
    temp = head
    while (temp):
        print(temp.getData(),end="->")
        temp = temp.getNext()
    

print("Linked list 1 : ",end = "")
traverse(head_1)

print()

print("Linked List 2 : ",end = "")
traverse(head_2)

print()

print("Intersection Linked List : ",end = "")



def intersection_lists(head1, head2):
    # Create a dummy node to serve as the head of the merged list
    dummy = Node(0)
    current = dummy
    l1 = []
    l2 = []
    intersection_list = []
    
    temp1 = head1
    temp2 = head2
    
    while temp1:
        l1.append(temp1.getData())
        temp1 = temp1.getNext()
            
    while temp2:
        l2.append(temp2.getData())
        temp2 = temp2.getNext()
        
    for i in l1:
        if i in l2:
            intersection_list.append(i)
    
    
    for i in intersection_list:
        newnode = Node(i)
        current.setNext(newnode)
        current = current.getNext()
        
    return dummy.getNext()

newhead = intersection_lists(head_1,head_2)
traverse(newhead)
```

8. Rotate a linked list by k positions to the right

   1->2->3->4->8->6->9 , after rotating for 2 times becomes , 3->4->8->6->9->1->2
    

```python
# Creating class Node
class Node:
    def __init__(self,Data,Next=None):
        self.__Data = Data
        self.__Next = Next
    
    def setData(self,Data):
        self.__Data = Data
        
    def getData(self):
        return self.__Data
    
    def setNext(self,Next):
        self.__Next = Next
        
    def getNext(self):
        return self.__Next

# Creating list nodes

head = Node(1)
node1 = Node(2)
node2 = Node(3)
node3 = Node(4)
node4 = Node(8)
node5 = Node(6)
node6 = Node(9)
        
    
# Creating a linked list

head.setNext(node1)
node1.setNext(node2)
node2.setNext(node3)
node3.setNext(node4)
node4.setNext(node5)
node5.setNext(node6)


# Traversing the linked list

def traverse(head):
    temp = head
    while (temp):
        print(temp.getData(),end="->")
        temp = temp.getNext()
    

# Calling the function for traversing the list        
print("Original Linked list : ")
traverse(head)

print()
print()

# Rotating Linked list by K positions
def rotate(head,k):
    # Traversing to the last node
    temp = head
    while(temp.getNext()):
        temp = temp.getNext()
        
    temp.setNext(head)
    
    i = 0
    while i < k :
        temp = temp.getNext()
        i += 1
        
    head = temp.getNext()
    temp.setNext(None)
    
    return head
    

# Calling the function for traversing the roated linked list        
print("Rotated Linked list : ")
newhead = rotate(head,2)
traverse(newhead)

```
